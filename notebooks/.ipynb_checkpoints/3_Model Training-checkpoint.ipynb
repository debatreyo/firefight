# Model Training & Evaluation

## Dataset
1. Regroups data recorded from two regions of Algeria-
   - Bejaia region, NE-Algeria
   - Sidi Bel-abbes region, NW-Algeria
2. Size- 244 instances (122 instances per region)
3. Contains 11 attributes/features and 1 target variable/class
4. 244 instances is split into two classes-
   - Fire (138 instances)
   - Not Fire (106 instances)
   
   
### Attributes/Features
|Attribute|Description|
|:---:|:---:|
|`Date`|Split into 3 columns- `day`, `month`, `year`|
|`Temperature`|Maximum noon temperature ($Â°C$)|
|`RH`|Relative humidity in $\%$|
|`Ws`|Wind speed in $km/h$|
|`Rain`|Total rainfall in $mm$|
|`FWI`|Fire Weather Index (FWI) Index|
|`FFMC`|Fine Fuel Moisture Code (FFMC) index from the FWI system|
|`DMC`|Duff Moisture Code (DMC) index from the FWI system|
|`DC`|Drought Code (DC) index from the FWI system|
|`ISI`|Initial Spread Index (ISI) from the FWI system|
|`BUI`|Buildup Index (BUI) from the FWI system|
|`Region`|Region from where data was recorded $\rightarrow$ Bejaia region, NE-Algeria (`0`) or Siddi-Bel Abbes Region, NW-Algeria (`1`)|
|`Classes`|Target variable. Two possible values- `fire`, `not fire`|

## Feature Selection

1. Correlation among the input features will be considered to drop similar features from final training pipeline.
2. Pearson Correlation coefficient will be used-

$$
\text{Pearson Correlation (r) = }
\frac{
    \sum_i
    (x_i - \bar x)
    \cdot
    (y_i - \bar y)
}{
    \sqrt{
        \sum_i
        (x_i - \bar x)^2
        \cdot
        \sum_i
        (y_i - \bar y)^2
    }
}
\\ \ \\
x_i = \text{Individual samples of first feature}\newline
\bar x = \text{Mean value of first feature}\newline
y_i = \text{Individual samples of second feature}\newline
\bar y = \text{Mean value of second feature}\newline
$$

#### Interpreting Correlation Coefficient Value
1. Value close to $+1$ $\rightarrow$ Large positive linear correlation (relationship)
2. Value close to $-1$ $\rightarrow$ Large negative linear correlation (relationship)
3. Value close to $0$ $\rightarrow$ No linear correlation (relationship)
    - the 2 variables (_most likely_) do not affect each other much
    
### Idea behind dropping features with high correlation
If two input features are highly correlated it signifies they are 'similar'. They behave similarly. When one changes by some magnitude, the other also behaves similarly.\
This raises issue of multicolinearity in the dataset.

> Multicolinearity: Presence of high correlations between two or more independent variables (predictors).

Multicolinearity is problematic because it becomes impossible to distinguish between the effects of the independent variables on the dependent variable.

Hence including both of them for predicting the target variable may be redundant.\
Only one of the two features will then be included in final training pipeline.

#### NOTE:
1. Correlation between the different input (independent) features are being considered only
2. We will not be considering the correlation between input features and target variable
    - if an input feature is highly correlated with the target (dependent feature) variable then it is fine

```python
# check for multicolinearity
multicolinearity = X_train.corr()
```

```python
# heatmap

fig, ax = plt.subplots(figsize=(7,7), dpi=220)

# customizations
annot_props = dict(size=7, family="times new roman")
text_props = dict(size=9, family="times new roman")

sns.heatmap(data=multicolinearity, annot=True, ax=ax,
           annot_kws=annot_props, cmap="Blues")

ax.set_xticklabels(labels=ax.get_xticklabels(), fontdict=text_props)
ax.set_yticklabels(labels=ax.get_yticklabels(), fontdict=text_props)

fig.tight_layout();
```

#### Helper function to select Input Features to drop

```python
def features_to_drop(dataset, threshold):
    """
    INPUT ->
    `dataset`: Pandas dataframe containing input features only
    `threshold`: Minimum threshold for pearson correlation coefficient
                between 2 input features.
                Features which have correlation > threshold will be returned.
    
    OUTPUT ->
    `cols_to_drop`: set of feature names which is to be dropped
                    to avoid multicolinearity.
    """
    # empty set to store names of columns to drop
    cols_to_drop = set()
    # correlation matrix
    correlation_matrix = dataset.corr()
    # column names (input features)
    col_names = correlation_matrix.columns
    
    # iterate through lower triangular (except diagonal) matrix
    # of the correlation matrix
    # NOTE: Correlation matrix is symmetric. Hence only checking lower triangle portion
    # will ensure all possible pairs of input features are considered.
    
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            # check correlation (pearson correlation)
            r = abs(correlation_matrix.iloc[i, j])
            # threshold check
            if r > threshold:
                cols_to_drop.add(col_names[i])
    
    # return set of column names (input features) to drop
    return cols_to_drop
```

```python
# threshold to drop columns
threshold = 0.85 ## if features have abs(correlation) > 0.85 then one of them will be dropped 

cols_to_drop = features_to_drop(X_train, threshold)

print(cols_to_drop)
```

```python
print(f"Before performing Feature Selection ->")
print(f"Total number of independent (input) features in dataset: {X_train.shape[1]}")

# drop columns from both Training and Test set
X_train = X_train.drop(columns=cols_to_drop)
X_test = X_test.drop(columns=cols_to_drop)

print(f"\nAfter performing Feature Selection ->\n")
print(f"Total number of independent (input) features in dataset: {X_train.shape[1]}")
```

## Feature Scaling (Standardization)
1. Z-score normalization will be performed on the dataset

$$
\text{Standardized Data = }
\frac{(X - \mu)}{\sigma}\\ \ \\
X: \text{Dataset matrix}\newline
\mu: \text{Mean of dataset}\newline
\sigma: \text{Standard Deviation of dataset}
$$

After feature scaling the input features will have a mean ($\mu$) of $0$ and standard deviation ($\sigma$) of $1$.

```python
# instantiate standard scalar object
scaler = StandardScaler()

# scale datasets

scaler.fit(X_train) ## fit to Training set
X_train_scaled = scaler.transform(X_train) ## scaling Training set
X_test_scaled = scaler.transform(X_test) ## scaling Test set
```

```python
print(f"First 5 rows of Scaled Training set ->\n\n {X_train_scaled[0:5, :]}")
```

#### Visualizing Impact of Standardizing the dataset

```python
fig, ax = plt.subplots(figsize=(9, 8), dpi=220,
                      nrows=2, ncols=1)

ax_unscaled, ax_scaled = ax

# customizations
text_props = dict(size=11, family="times new roman")
text_props_title = dict(size=13, family="times new roman")

# unscaled dataset
sns.boxplot(X_train, ax=ax_unscaled)
ax_unscaled.set_title("Unscaled Training set", fontdict=text_props_title)
ax_unscaled.set_xticks(ax_unscaled.get_xticks(), labels=X_train.columns,
                      fontdict=text_props)

# scaled dataset
sns.boxplot(X_train_scaled, ax=ax_scaled)
ax_scaled.set_title("Training set (Standardized)", fontdict=text_props_title)
ax_scaled.set_xticks(ax_scaled.get_xticks(), labels=X_train.columns,
                    fontdict=text_props)


fig.tight_layout();
```

## Machine Learning Model building

#### Helper Functions

```python
def plot_model(y_test, y_pred, model_name=None):
    """
    
    """
    fig, ax = plt.subplots(figsize=(9,4), dpi=220,
                      nrows=1, ncols=2)
    ax_scatter, ax_kde = ax
    
    # Scatter plot
    ax_scatter.scatter(y_test, y_pred)
    ax_scatter.set_title("Scatter Plot")
    ax_scatter.set_xlabel("True Value")
    ax_scatter.set_ylabel("Prediction")
    ax_scatter.plot(y_test, y_test, label="Perfect fit line", color="red")
    ax_scatter.legend()

    sns.kdeplot(data=y_test, label="True Values", ax=ax_kde)
    sns.kdeplot(data=y_pred, label="Predicted Values", ax=ax_kde)
    ax_kde.set_title("Distribution Plot")
    ax_kde.legend()
    
    if model_name is None:
        fig.suptitle(f"Model Performance", fontsize=15)
    else:
        fig.suptitle(f"{model_name.title()} Model Performance", fontsize=15)
        
    fig.tight_layout();
```

```python
def train_predict(model, X_train, y_train, X_test):
    """
    
    """
    # fit model to scaled Training set
    model.fit(X_train, y_train)

    # make predictions on Test set with the trained model
    predictions = model.predict(X_test)
    
    return predictions
```

```python
def evaluate(model, y_test, predictions, model_name=None):
    """
    
    """
    # mean absolute error
    mae = mean_absolute_error(y_true=y_test, y_pred=predictions)
    # coefficient of determination
    r2 = r2_score(y_true=y_test, y_pred=predictions)
    
    if model_name is None:
        print(f"Model Performance ->\n")
        print(f"Mean Absolute Error: {mae}")
        print(f"Coefficient of Determination (R^2): {r2}")
    else:
        print(f"{model_name} Model Performance ->\n")
        print(f"Mean Absolute Error: {mae}")
        print(f"Coefficient of Determination (R^2): {r2}")
    
    return mae, r2
```

### Linear Regression Model

```python
# for model building and training
from sklearn.linear_model import LinearRegression
```

```python
# instantiate linear regression object
lin_reg = LinearRegression()

# train & make predictions on Test set with the trained model
y_pred_lin_reg = train_predict(lin_reg, X_train_scaled, y_train, X_test_scaled)
```

```python
# evaluate model
mae_lin_reg, r2_lin_reg = evaluate(lin_reg, y_test, y_pred_lin_reg, model_name="Linear Regression")
```

```python
# visualize model performance
plot_model(y_test, y_pred_lin_reg, "Linear Regression")
```

```python
# instantiate model
lasso = Lasso()

# train & make predictions
y_pred_lasso = train_predict(lasso, X_train_scaled, y_train, X_test_scaled)
```

```python
# evaluate model
mae_lasso, r2_lasso = evaluate(lasso, y_test, y_pred_lasso, model_name="Lasso Regression")
```
