# Exploratory Data Analysis
---

## Dataset
1. Regroups data recorded from two regions of Algeria-
   - Bejaia region, NE-Algeria
   - Sidi Bel-abbes region, NW-Algeria
2. Size- 244 instances (122 instances per region)
3. Contains 11 attributes/features and 1 target variable/class
4. 244 instances is split into two classes-
   - Fire (138 instances)
   - Not Fire (106 instances)
   
   
### Attributes/Features
|Attribute|Description|
|:---:|:---:|
|`Date`|Split into 3 columns- `day`, `month`, `year`|
|`Temperature`|Maximum noon temperature ($Â°C$)|
|`RH`|Relative humidity in $\%$|
|`Ws`|Wind speed in $km/h$|
|`Rain`|Total rainfall in $mm$|
|`FWI`|Fire Weather Index (FWI) Index|
|`FFMC`|Fine Fuel Moisture Code (FFMC) index from the FWI system|
|`DMC`|Duff Moisture Code (DMC) index from the FWI system|
|`DC`|Drought Code (DC) index from the FWI system|
|`ISI`|Initial Spread Index (ISI) from the FWI system|
|`BUI`|Buildup Index (BUI) from the FWI system|
|`Classes`|Target variable. Two possible values- `fire`, `not fire`|

## Exploring The Data further

### Is the dataset balanced?

````python
# check `classes` target variable
df_eda["Classes"].value_counts()
````

````python
# one-hot-encode target variable
condition = df_eda["Classes"].str.contains(pat=".*(not).*", regex=True)

df_eda["Classes"] = np.where(condition, 0, 1) ## if condition is true then set it to 0 (not fire), else 1
````

````python
# check dataset
df_eda.head()
````

````python
# check target variable
df_eda["Classes"].value_counts()
````

````python
df_eda["Classes"].value_counts().values
````

````python
# check dataset for imbalance
fig, ax = plt.subplots(figsize=(3,3), dpi=220)

## data
vals = df_eda["Classes"].value_counts().values
labels = list(map(lambda class_: "Fire" if class_==1 else "Not Fire",
                  df_eda["Classes"].value_counts().index))

## customizations
colors = ["#A1DD70", "#FF6969"]
wedge_props = dict(linewidth=0.25, edgecolor="black")

ax.pie(x=vals, labels=None, colors=colors,
       wedgeprops=wedge_props, autopct="%1.1f%%",
      pctdistance=1.13, radius=0.95)

ax.legend(labels, loc="center left",
         bbox_to_anchor=(1, 0.5))


fig.tight_layout();
````

#### Observations
1. Dataset is not perfectly balanced, however not too unbalanced either.
    - $\approx 56\%$ instances of positive class against $\approx 43\%$ negative

### How are the input features Correlated?

```python
def custom_regplot(x_col, y_col, dataframe):
    """
    Creates a Seaborn Regression plot.
    Arguments ->
    x_col: (str) Name of column/variable/feature to use for X-axis
    y_col: (str) Name of column/variable/feature to use for Y-axis
    dataframe: (Pandas DataFrame) Dataframe containing
                dataset having these columns/variables/features
                
    """
    # figure & axes objects
    fig, ax = plt.subplots(figsize=(6,4), dpi=250)
    # plot regression plot on above created axes object
    sns.regplot(x=x_col, y=y_col, data=dataframe,
          line_kws={"linewidth":0.8, "linestyle":"--", "color":"#00224D"},
          scatter_kws={"marker":"o", "edgecolors":"#114232", "alpha":0.6,
"color":"#4CCD99", "linewidth":0.45, "s":10},
           ax=ax)
    
    # adjust tickmarks
    ax.set_xticks(ticks=ax.get_xticks(), fontsize=8)
    ax.set_yticks(ticks=ax.get_yticks(), fontsize=8)

    fig.tight_layout();
```

### How do the two regions compare?

1. Two regions-
    - `0`: Bejaia region, NE-Algeria
    - `1`: Siddi-Bel Abbes Region, NW-Algeria

````python
# original dataframe with dates
df_bejaia = dataset[dataset["Region"] == 0].copy()
df_siddi = dataset[dataset["Region"] == 1].copy()
````

````python
# encode the target variable in both dataframes
condition_bejaia = df_bejaia["Classes"].str.contains(pat=".*(not).*", regex=True)
condition_siddi = df_siddi["Classes"].str.contains(pat=".*(not).*", regex=True)

df_bejaia["Classes"] = np.where(condition_bejaia, "Not Fire", "Fire")
df_siddi["Classes"] = np.where(condition_siddi, "Not Fire", "Fire")
````

````python
# plot number of fires over months

fig, ax = plt.subplots(figsize=(15,6), nrows=1, ncols=2)
ax_bejaia, ax_siddi = ax

# customizations
text_props = dict(size=18, family="times new roman")

month_labels = ["June", "July", "August", "September"]

sns.countplot(x="month", hue="Classes", data=df_bejaia, ax=ax_bejaia)
sns.countplot(x="month", hue="Classes", data=df_siddi, ax=ax_siddi)

ax_bejaia.set_title("Bejaia Region, NE-Algeria", fontdict=text_props)
ax_bejaia.set_xticks(ax_bejaia.get_xticks(), labels=month_labels, fontdict=text_props)
ax_bejaia.set_xlabel("Month", fontdict=text_props)
ax_bejaia.set_ylabel("Count", fontdict=text_props)




ax_siddi.set_title("Siddi-Bel Abbes Region, NW-Algeria", fontdict=text_props)
ax_siddi.set_xticks(ax_siddi.get_xticks(), labels=month_labels, fontdict=text_props)
ax_siddi.set_xlabel("Month", fontdict=text_props)
ax_siddi.set_ylabel("Count", fontdict=text_props)


fig.tight_layout();
````
